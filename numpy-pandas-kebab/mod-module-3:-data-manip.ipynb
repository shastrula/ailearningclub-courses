## Key Concept: Series and DataFrames

A fundamental data structure in Pandas is the Series, which represents a one-dimensional labeled array. You can also create DataFrames, two-dimensional data structures that resemble tables or spreadsheets.

```python
# Importing necessary libraries
import pandas as pd

# Creating a simple Series
data = [28, 24]
s = pd.Series(data, index=['John', 'Anna'], name='Age')
print(s)
# Output:
# John    28
# Anna    24
# Name: Age, dtype: int64
```
