# Module 1: Introduction to Numpy and Pandas

**Duration:** 5 min

Welcome to the Numpy and Pandas course. In this module, we introduce the two most foundational libraries for data science in Python.

## Why Numpy and Pandas?

Numpy provides support for fast, multi-dimensional arrays and mathematical operations. Pandas builds on Numpy to provide easy-to-use data structures like DataFrames for data manipulation.

> **💡 Tip:** It is highly recommended to learn Numpy arrays first before diving into Pandas DataFrames, as Pandas is built directly on top of Numpy.
