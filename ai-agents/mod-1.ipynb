## ReAct Framework

The ReAct (Reason and Act) framework is a methodology where an AI agent reasons about a task and then performs actions based on that reasoning. This approach allows agents to handle complex tasks by breaking them down into smaller, manageable steps.

````python
def reason(task):
    """Simulate reasoning about a task."""
    plan = f'Plan to execute: {task}'
    return plan

def act(plan):
    """Simulate acting based on the plan."""
    print(f'Executing: {plan}')

# Example usage
task = 'fetch data from API'
plan = reason(task)
act(plan)
````

## LangGraph for Multi-Agent Systems

LangGraph is a library for creating multi-agent systems where multiple AI agents collaborate to achieve a common goal. This approach leverages the strengths of individual agents to solve more complex problems.

````python
from langgraph import MultiAgentSystem, Agent

# Define agents
agent1 = Agent('Agent1', lambda task: f'Agent1 processes: {task}')
agent2 = Agent('Agent2', lambda task: f'Agent2 processes: {task}')

# Create multi-agent system
mas = MultiAgentSystem([agent1, agent2])

# Execute task
task = 'complex task'
result = mas.execute(task)
print(result)
````

> **💡 Tip:** Ensure that each agent in a multi-agent system has a clearly defined role to avoid redundancy and improve efficiency.
