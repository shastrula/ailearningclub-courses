# Understanding ReAct Framework

**Duration:** 5 min

This module delves into the ReAct (Reason + Act) framework, an essential tool for developing AI agents that can reason about tasks and act upon them. Understanding this framework is crucial for creating sophisticated AI systems that can handle complex workflows and decision-making processes.

## Reasoning in ReAct Framework

The reasoning component of the ReAct framework allows AI agents to analyze and understand tasks. It involves breaking down a task into smaller, manageable parts and determining the necessary actions to complete it. This component is vital for ensuring that the agent can make informed decisions based on the given context.

````python
def reason(task):
    """Analyze the task and determine the necessary actions."""
    actions = []
    if 'sort' in task:
        actions.append('sorting')
    if 'filter' in task:
        actions.append('filtering')
    return actions

# Example usage
task ='sort the list of numbers'
print(reason(task))
````

## Acting in ReAct Framework

The acting component of the ReAct framework involves executing the actions determined by the reasoning component. This can include performing calculations, manipulating data, or interacting with external systems. The acting component ensures that the AI agent can take concrete steps to achieve its goals.

````python
def act(actions, data):
    """Execute the determined actions on the given data."""
    result = data
    for action in actions:
        if action =='sorting':
            result.sort()
        elif action == 'filtering':
            result = [x for x in result if x > 0]
    return result

# Example usage
data = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
actions = ['sorting', 'filtering']
print(act(actions, data))
````

> **💡 Tip:** Ensure that the actions determined by the reasoning component are feasible and relevant to the task to avoid unnecessary computations or errors.
