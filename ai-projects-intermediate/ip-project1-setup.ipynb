# Project 1 — Sentiment Analysis: Problem & Data

**Duration:** 5 min

## Problem Statement

Classify Amazon product reviews as positive (rating ≥ 4) or negative (rating ≤ 2). You'll fine-tune a pre-trained DistilBERT model using HuggingFace Transformers.

## Dataset

> **💡 Tip:** 📦 Dataset: Amazon Polarity (HuggingFace Datasets)
from datasets import load_dataset
dataset = load_dataset('amazon_polarity')

Or use the smaller version:
https://www.kaggle.com/datasets/bittlingmayer/amazonreviews

3.6M reviews, binary labels. We'll use a 10k subset for training speed.

## Deliverables

• Fine-tune DistilBERT on 10k Amazon reviews
• Achieve > 92% accuracy on validation set
• Build an inference function: predict_sentiment('review text') → positive/negative + confidence
• Plot training loss curve
