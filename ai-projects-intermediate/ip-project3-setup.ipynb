# Project 3 — Movie Recommender: Problem & Data

**Duration:** 5 min

## Problem Statement

Build a collaborative filtering recommender that predicts movie ratings and generates personalised recommendations for any user. Dataset: MovieLens 100K (100,000 ratings from 943 users on 1,682 movies).

## Dataset

> **💡 Tip:** 📦 Dataset: MovieLens 100K
https://grouplens.org/datasets/movielens/100k/

Direct download: https://files.grouplens.org/datasets/movielens/ml-100k.zip

Key file: u.data — userId, movieId, rating, timestamp

## Deliverables

• Build a matrix factorization model using PyTorch embeddings
• Achieve RMSE < 0.95 on test set
• Build recommend(user_id, n=10) function returning top-N movie titles
• Compare with a simple baseline (user mean rating)
