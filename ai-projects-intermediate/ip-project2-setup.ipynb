# Project 2 — Image Classifier: Problem & Data

**Duration:** 5 min

## Problem Statement

Build a CNN to classify images into 10 categories: airplane, automobile, bird, cat, deer, dog, frog, horse, ship, truck. Dataset: CIFAR-10 (60,000 32×32 colour images).

## Dataset

> **💡 Tip:** 📦 Dataset: CIFAR-10 — built into PyTorch/Keras

PyTorch: torchvision.datasets.CIFAR10
Keras:   keras.datasets.cifar10

No download needed. 50k training images, 10k test images.

## Deliverables

• Build a CNN with at least 3 conv layers + batch norm + dropout
• Achieve > 80% test accuracy
• Plot training/validation accuracy curves
• Visualize 10 misclassified images with predicted vs true labels
• Save model weights
