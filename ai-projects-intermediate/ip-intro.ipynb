# Intermediate AI Projects Overview

**Duration:** 5 min

Apply machine learning to real-world datasets and build end-to-end projects.

## Project Goals

Master the complete ML workflow by building three intermediate projects:

1. **Customer Churn Prediction** - Predict which customers will leave
2. **Sentiment Analysis** - Classify text as positive, negative, or neutral
3. **Image Classification** - Build a model to recognize objects in images

## What You'll Learn

- Data exploration and visualization
- Feature engineering and preprocessing
- Model selection and hyperparameter tuning
- Cross-validation and performance metrics
- Saving and loading trained models

## Project Workflow

```
Data Collection → Exploration → Preprocessing → 
Feature Engineering → Model Training → Evaluation → 
Optimization → Deployment
```

> **Try it in Google Colab:** [![Open in Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/shastrula/ailearningclub-courses/blob/main/ai-projects-intermediate/ip-intro.ipynb)


## Tools You'll Use

- **Pandas** - Data manipulation
- **Scikit-learn** - Machine learning models
- **Matplotlib/Seaborn** - Visualization
- **Jupyter Notebooks** - Interactive development

> **💡 Tip:** Start with customer churn prediction as it's the most straightforward. Build confidence before moving to NLP and computer vision projects.
