# Project 3 — Iris Classifier: Problem & Data

**Duration:** 5 min

## Problem Statement

Classify iris flowers into 3 species (Setosa, Versicolor, Virginica) using 4 measurements: sepal length, sepal width, petal length, petal width. This is a multiclass classification problem.

## Dataset

> **💡 Tip:** 📦 Dataset: Iris — built into scikit-learn
from sklearn.datasets import load_iris

No download needed! Also available at:
https://archive.ics.uci.edu/dataset/53/iris

## Deliverables

• Load and visualize the dataset (pairplot)
• Train 3 different classifiers and compare accuracy
• Plot a confusion matrix
• Save the best model with joblib
• Write a predict() function that takes 4 measurements and returns the species name
