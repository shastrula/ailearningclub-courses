# Project 1 — Titanic: Problem & Data

**Duration:** 5 min

## Problem Statement

The RMS Titanic sank in 1912. Using passenger data (age, sex, class, fare, cabin), predict who survived. This is a binary classification problem: survived (1) or not (0).

## Dataset

> **💡 Tip:** 📦 Dataset: Titanic — Machine Learning from Disaster
https://www.kaggle.com/c/titanic/data

Download train.csv and test.csv. Or open the Colab notebook below — it loads the data automatically.

## Open the Starter Notebook

> **💡 Tip:** 🚀 Colab Notebook: https://colab.research.google.com/github/datasciencedojo/datasets/blob/master/titanic_starter.ipynb

If the link above is unavailable, create a new Colab notebook and follow the steps in the next modules.

## Deliverables

By the end of this project you will have:
• Loaded and explored the Titanic dataset
• Handled missing values (Age, Cabin, Embarked)
• Engineered at least 2 new features
• Trained a classification model (Logistic Regression or Random Forest)
• Achieved > 78% accuracy on the validation set
