# How Live Projects Work

**Duration:** 5 min

## Welcome to Live Projects

Each project gives you a real dataset, a problem statement, a starter notebook, and clear deliverables. You work through it step by step, then mark it complete to earn progress toward your certificate.

> **💡 Tip:** Run notebooks in Google Colab — free GPU, no setup. Click the Colab badge in each project to open it instantly.

## What You'll Build

Project 1: Titanic Survival Predictor — binary classification with real passenger data.
Project 2: House Price Estimator — regression with the Ames Housing dataset.
Project 3: Iris Flower Classifier — multiclass classification, your first end-to-end ML pipeline.
