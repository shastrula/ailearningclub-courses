# Project 2 — House Prices: Problem & Data

**Duration:** 5 min

## Problem Statement

Predict the sale price of houses in Ames, Iowa using 79 features including lot size, neighborhood, year built, and quality ratings. This is a regression problem — your output is a continuous price value.

## Dataset

> **💡 Tip:** 📦 Dataset: House Prices — Advanced Regression Techniques
https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data

Download train.csv. Target column: SalePrice.

## Deliverables

• Handle missing values across 79 features
• Engineer at least 3 new features (TotalSF, HouseAge, etc.)
• Train a gradient boosting model (XGBoost or LightGBM)
• Achieve RMSE < 25,000 on validation set
• Plot predicted vs actual prices
