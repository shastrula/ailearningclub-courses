# Project 1 — RAG Pipeline: Problem & Data

**Duration:** 5 min

## Problem Statement

Build a Retrieval-Augmented Generation (RAG) system that answers questions about AI research papers. Users ask natural language questions; the system retrieves relevant chunks and generates grounded answers.

## Dataset

> **💡 Tip:** 📦 Dataset: AI ArXiv Papers (HuggingFace)
from datasets import load_dataset
dataset = load_dataset('scientific_papers', 'arxiv', split='train[:500]')

Or use your own PDFs — the pipeline works with any text documents.

## Deliverables

• Chunk and embed 500 ArXiv abstracts into a FAISS vector store
• Build a retrieval chain: query → top-5 chunks → LLM answer
• Evaluate with 3 test questions and show retrieved sources
• Add a simple Gradio UI for interactive Q&A
