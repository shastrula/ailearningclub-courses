# Project 2 — LLM Fine-tuning: Problem & Data

**Duration:** 5 min

## Problem Statement

Fine-tune a small LLM to generate SQL queries from natural language questions. Given 'Show me all customers who spent more than $500', the model should output the correct SQL.

## Dataset

> **💡 Tip:** 📦 Dataset: Spider SQL (HuggingFace)
from datasets import load_dataset
dataset = load_dataset('spider')

7,000 training examples of (question, SQL) pairs across 200 databases.
https://huggingface.co/datasets/spider

## Deliverables

• Fine-tune Phi-2 (2.7B) or TinyLlama (1.1B) using QLoRA (4-bit quantization)
• Train for 1 epoch on Spider train split (~7k examples)
• Evaluate on 100 test examples — measure exact match accuracy
• Save LoRA adapter weights
• Build text2sql(question, schema) inference function
