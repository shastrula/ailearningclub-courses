# Project 3 — Fraud Detection: Problem & Data

**Duration:** 5 min

## Problem Statement

Build a real-time credit card fraud detection system. The dataset is highly imbalanced (0.17% fraud). You'll handle imbalance, tune the decision threshold, and serve predictions via a FastAPI endpoint.

## Dataset

> **💡 Tip:** 📦 Dataset: Credit Card Fraud Detection
https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

284,807 transactions, 492 frauds (0.17%). Features are PCA-transformed (V1-V28) + Amount + Time.
Direct download requires Kaggle account. Alternatively:
https://huggingface.co/datasets/pointe77/credit-card-fraud-detection

## Deliverables

• Handle class imbalance with SMOTE oversampling
• Train XGBoost classifier
• Tune decision threshold to maximise F1 on fraud class
• Achieve Precision > 0.85 and Recall > 0.80 on fraud class
• Serve predictions via FastAPI: POST /predict → {fraud: bool, probability: float}
