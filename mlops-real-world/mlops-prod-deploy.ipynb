# PROD — Deploy to AWS

**Duration:** 5 min

## Option A: AWS Lambda (serverless, cheapest)

```python
# Package as Lambda container image
aws ecr create-repository --repository-name churn-api --region us-east-1

# Create Lambda function from container
aws lambda create-function \
  --function-name churn-prediction \
  --package-type Image \
  --code ImageUri=<account>.dkr.ecr.us-east-1.amazonaws.com/churn-api:latest \
  --role arn:aws:iam::<account>:role/lambda-execution-role \
  --timeout 30 --memory-size 512 --region us-east-1

# Add API Gateway trigger via console or:
aws apigatewayv2 create-api --name churn-api --protocol-type HTTP
```

## Option B: EC2 (always-on, more control)

```python
# On EC2 (t3.small, ~$15/month)
sudo apt-get update && sudo apt-get install -y docker.io
sudo docker pull <account>.dkr.ecr.us-east-1.amazonaws.com/churn-api:latest
sudo docker run -d -p 80:8000 --restart always churn-api:latest
```

> **💡 Tip:** For a side project, Lambda is the right choice — you pay only per request (~$0.20 per 1M requests). EC2 makes sense when you need persistent connections or > 15 min processing.
