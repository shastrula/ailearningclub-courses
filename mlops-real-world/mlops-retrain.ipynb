# PROD — Automated Retraining Pipeline

**Duration:** 5 min

## Retraining Trigger

When drift is detected, the pipeline should: collect new labeled data → retrain → evaluate → if better than current model → deploy automatically. This is the core of MLOps.

```python
name: Retrain on Drift

on:
  workflow_dispatch:  # manual trigger
  schedule:
    - cron: '0 2 * * 1'  # every Monday at 2am

jobs:
  retrain:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - uses: actions/setup-python@v5
        with: { python-version: '3.11' }
      - run: pip install -r requirements.txt
      - name: Run drift check
        run: python monitor/drift_check.py
      - name: Retrain model
        run: python src/train.py
      - name: Run tests
        run: pytest tests/ -v
      - name: Deploy if tests pass
        run: |
          docker build -t churn-api:retrained .
          # push and update Lambda/EC2
```
