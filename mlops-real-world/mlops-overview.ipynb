# Project Overview & Stack

**Duration:** 5 min

## What You'll Build

A customer churn prediction service — end to end:

Dev → train model, track experiments with MLflow
QA → containerize with Docker, write integration tests, CI with GitHub Actions
Prod → deploy to AWS Lambda (or EC2), monitor with CloudWatch, set up drift alerts

## Stack

> **💡 Tip:** Tools used:
• scikit-learn — model training
• MLflow — experiment tracking
• Docker — containerization
• GitHub Actions — CI/CD pipeline
• FastAPI — serving endpoint
• AWS Lambda or EC2 — production hosting
• CloudWatch — monitoring & alerting
• Evidently AI — data drift detection

## Dataset

> **💡 Tip:** 📦 Dataset: Telco Customer Churn
https://www.kaggle.com/datasets/blastchar/telco-customer-churn

7,043 customers, 21 features, binary target: Churn (Yes/No).
Also available: https://huggingface.co/datasets/scikit-learn/churn-prediction
