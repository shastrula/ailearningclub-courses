## Low-Rank Adaptation (LoRA)

LoRA is a technique that allows for efficient fine-tuning of LLMs by introducing low-rank matrices to adapt the model parameters. This method reduces the number of trainable parameters, making the fine-tuning process more memory-efficient and faster. LoRA has been widely adopted due to its effectiveness in adapting pre-trained models to new tasks with minimal computational overhead.

````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load pre-trained model and tokenizer
model_name = 'EleutherAI/gpt-neo-125M'
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Define LoRA parameters
lora_r = 8  # Rank of decomposition
lora_alpha = 32  # Alpha parameter for LoRA
lora_dropout = 0.1  # Dropout probability for LoRA

# Apply LoRA to the model
for name, param in model.named_parameters():
    if 'query_key_value' in name:  # Target the query, key, and value matrices
        param.data += torch.randn_like(param) * lora_alpha / lora_r

print('LoRA applied to the model.')
````

## Quantized Low-Rank Adaptation (QLoRA)

QLoRA extends the LoRA technique by incorporating quantization, which further reduces the memory footprint and computational cost of fine-tuning LLMs. Quantization involves converting the model parameters to lower precision, such as int8, which allows for more efficient storage and faster inference. QLoRA is particularly useful for deploying LLMs on resource-constrained devices.

````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load pre-trained model and tokenizer
model_name = 'EleutherAI/gpt-neo-125M'
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Define QLoRA parameters
lora_r = 8  # Rank of decomposition
lora_alpha = 32  # Alpha parameter for LoRA
lora_dropout = 0.1  # Dropout probability for LoRA
quantization_bits = 8  # Quantization bits

# Apply QLoRA to the model
for name, param in model.named_parameters():
    if 'query_key_value' in name:  # Target the query, key, and value matrices
        param.data = torch.quantize_per_tensor(param.data, scale=1.0, zero_point=0, dtype=torch.quint8)
        param.data += torch.randn_like(param) * lora_alpha / lora_r

print('QLoRA applied to the model.')
````

> **💡 Tip:** When applying LoRA or QLoRA, ensure that the rank 'r' is chosen appropriately to balance between memory efficiency and model performance. A too-low rank may not capture sufficient information, while a too-high rank may negate the benefits of these techniques.
