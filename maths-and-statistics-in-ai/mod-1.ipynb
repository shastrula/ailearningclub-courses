## Linear Algebra

Linear algebra is fundamental in AI, particularly for operations involving vectors and matrices. It allows for efficient computation of large datasets, which is common in machine learning. Key concepts include vector addition, scalar multiplication, matrix multiplication, and eigenvalues and eigenvectors.

````python
import numpy as np

# Vector addition
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Scalar multiplication
c = 2 * a

# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = np.dot(A, B)

# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print('Vector Addition:', a + b)
print('Scalar Multiplication:', c)
print('Matrix Multiplication:'
, C)
print('Eigenvalues:', eigenvalues)
print('Eigenvectors:', eigenvectors)
````

## Calculus

Calculus is vital for understanding the optimization of functions, which is a core part of training machine learning models. Concepts such as derivatives and integrals are used to minimize loss functions and improve model accuracy. Gradient descent, a common optimization algorithm, relies heavily on these principles.

````python
import sympy as sp

x = sp.symbols('x')

# Derivative of a function
f = x**2 + 2*x + 1
f_prime = sp.diff(f, x)

# Integral of a function
f_integral = sp.integrate(f, x)

print('Derivative:', f_prime)
print('Integral:', f_integral)
````

> **💡 Tip:** When using calculus in optimization, ensure your functions are differentiable to avoid issues with gradient descent.
